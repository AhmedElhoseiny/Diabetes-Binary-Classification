```python
x_data = df.iloc[:,:-1]
y_data = df.iloc[:,-1]
```

```python
x_data
```

```python
y_data
```

```python
from sklearn.preprocessing import StandardScaler

scaler_StandardScaler = StandardScaler()
X_scaled_StandardScaler = scaler_StandardScaler.fit_transform(x_data)
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x_data,y_data,test_size=0.2,random_state=1) 
```

```python
x_train_selected_Standardfcaler = scaler_StandardScaler.fit_transform(x_train)
x_test_selected_Standardfcaler = scaler_StandardScaler.transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB

GaussianNB_classifier = GaussianNB()

GaussianNB_classifier.fit(x_train_selected_Standardfcaler, y_train)
```

```python
y_pred_Standardfcaler = GaussianNB_classifier.predict(x_test_selected_Standardfcaler)
```

```python
from sklearn.metrics import accuracy_score

accuracy_Standardfcaler = accuracy_score(y_test, y_pred_Standardfcaler) * 100
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(2, 4))
plt.bar(['Standardfcaler'], [accuracy_Standardfcaler], color=['blue'])
plt.title('Sentiment Analysis Accuracy')
plt.show()

print(f'accuracy Standardfcaler:{accuracy_Standardfcaler} %')
```

```python
from sklearn.svm import SVC

svm_classifier = SVC()

svm_classifier.fit(x_train_selected_Standardfcaler, y_train)
```

```python
y_pred_Standardfcaler_s = svm_classifier.predict(x_test_selected_Standardfcaler)
```

```python
from sklearn.metrics import accuracy_score

accuracy_Standardfcaler = accuracy_score(y_test, y_pred_Standardfcaler_s) * 100
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(2, 4))
plt.bar(['Standardfcaler'], [accuracy_Standardfcaler], color=['red'])
plt.title('Sentiment Analysis Accuracy')
plt.show()

print(f'accuracy Standardfcaler:{accuracy_Standardfcaler} %')
```
